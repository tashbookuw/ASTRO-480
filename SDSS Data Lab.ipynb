```python
xmatch = QTable.read('xmatch.csv',format='ascii')
xmatch
```

```python
n_ra = []
n_dec = []
for i in range(0, len(ra)):
    if (ra[i] > 132.82 and ra[i] < 132.88) and (dec[i] > 11.768 and dec[i] < 11.83 ):
        n_ra.append(ra[i])
        n_dec.append(dec[i])
```

```python
#plt.plot(xmatch['ra'], xmatch['dec'], ra, dec 'r.')
#plt.plot(ra, dec, 'y.')
fig = plt.figure()
ax = fig.add_subplot(111)

plt.plot(xmatch['ra'], xmatch['dec'], 'ro', markersize=22, alpha=0.3)
plt.plot(n_ra, n_dec, 'yo', markersize=22, alpha=0.3)
#ax.plot(ra, dec, 'y.')
plt.show()
```

```python
rb_w = []
db_w = []
rb_s = []
db_s = []
whatsclose = .000000001
for i in range(0, len(n_ra)):
    for j in range(0, len(xmatch['ra'])):
        if n_ra[i] > (xmatch['ra'][j]-whatsclose) and n_ra[i] < (xmatch['ra'][j]+whatsclose) and n_dec[i] > (xmatch['dec'][j]-whatsclose) and n_dec[i] < (xmatch['dec'][j]+whatsclose) :
            rb_w.append(xmatch['ra'][j])
            db_w.append(xmatch['dec'][j])
            rb_s.append(n_ra[i])
            db_s.append(n_dec[i])
```

```python
fig, ax = plt.subplots(figsize=(20, 10))

plt.plot(rb_w, rb_w, 'ro', markersize=20, alpha=0.1)
plt.plot(rb_s, rb_s, 'bo', markersize=20, alpha=0.1)
#ax.plot(ra, dec, 'y.')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 10))

plt.plot(db_w, db_w, 'ro', markersize=20, alpha=0.1)
plt.plot(db_s, db_s, 'bo', markersize=20, alpha=0.1)
#ax.plot(ra, dec, 'y.')
plt.show()
```
